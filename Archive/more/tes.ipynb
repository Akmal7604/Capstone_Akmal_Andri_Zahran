```python
df.dropna(subset=["RecipeId", "Name", "AuthorId", "AuthorName", "CookTime", "PrepTime", "TotalTime",
    "DatePublished", "Description", "Images", "RecipeCategory", "Keywords",
    "RecipeIngredientQuantities", "RecipeIngredientParts", "AggregatedRating", "ReviewCount",
    "Calories", "FatContent", "SaturatedFatContent", "CholesterolContent", "SodiumContent",
    "CarbohydrateContent", "FiberContent", "SugarContent", "ProteinContent", "RecipeServings", "RecipeInstructions"
], inplace=True)
```

```python
df.drop(columns=['RecipeYield'], inplace=True)
df.info()
```

```python
# Menghapus baris yang memiliki nilai 0 di kolom tertentu
cols_to_check = [
    'Calories',
    'FatContent',
    'SaturatedFatContent',
    'CholesterolContent',
    'SodiumContent',
    'CarbohydrateContent',
    'FiberContent',
    'SugarContent',
    'ProteinContent'
]

# Filter DataFrame, hanya ambil baris yang tidak memiliki nilai 0 pada kolom yang ditentukan
df = df[(df[cols_to_check] != 0).all(axis=1)]

# Menghapus baris yang memiliki Images empty string
df = df[df['Images'] != "character(0)"]

# Menampilkan DataFrame yang telah dibersihkan
print(df)
```

```python
df.info()
df.head()
```

```python
# Fungsi untuk menghapus outlier berdasarkan IQR
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 3 * IQR
    upper_bound = Q3 + 3 * IQR
    
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Menerapkan fungsi ke setiap kolom
for col in cols_to_check:
    df = remove_outliers_iqr(df, col)

# Menampilkan DataFrame yang telah dibersihkan dari outlier
print(df)

# Pastikan untuk memeriksa perubahan dalam jumlah baris
print("Jumlah baris setelah penghapusan outlier:", df.shape[0])
```

```python
# Membulatkan dan mengubah tipe data menjadi integer untuk setiap kolom dalam list
for col in cols_to_check:
    df[col] = df[col].round().astype(int)

# Menampilkan DataFrame yang telah dibulatkan
print(df)

# Jika Anda ingin memeriksa hasilnya
print("Data setelah pembulatan:")
```

```python
import pandas as pd

# Misalkan df adalah DataFrame yang Anda miliki
# Contoh: df = pd.DataFrame({...})

# Menentukan kolom yang ingin dibersihkan
cols_to_clean = [
    'Images',
    'Keywords',
    'RecipeIngredientQuantities',
    'RecipeIngredientParts',
    'RecipeInstructions'
]

# Menghapus 'c()' serta tanda kutip ganda dari setiap kolom yang ditentukan
for col in cols_to_clean:
    df[col] = df[col].str.replace(r'c\(\s*|\s*\)|\"', '', regex=True)

# Menampilkan DataFrame yang telah dibersihkan
print(df)

# Jika Anda ingin memeriksa hasilnya
print("Data setelah pembersihan:")
print(df[cols_to_clean].head())  # Menampilkan beberapa baris pertama dari kolom yang dibersihkan
```

```python
# Convert time formats
def convert_time(time_str):
    if pd.isna(time_str):
        return 0
    try:
        time_str = str(time_str).upper().replace('PT', '')
        total = 0
        if 'H' in time_str:
            h_part = time_str.split('H')[0]
            total += int(h_part) * 60
            time_str = time_str.split('H')[1]
        if 'M' in time_str:
            m_part = time_str.split('M')[0]
            total += int(m_part) if m_part else 0
        return total
    except:
        return 0

for col in ['CookTime', 'PrepTime', 'TotalTime']:
    df[col] = df[col].apply(convert_time)
        
        
```

```python
# Mengambil semua kategori resep dari kolom 'RecipeCategory'
all_categories = df['RecipeCategory'].str.cat(sep=',').split(',')

# Menghapus spasi di awal dan akhir, kemudian mengambil unique values
unique_categories = set(category.strip() for category in all_categories)

# Menampilkan hasil
print(unique_categories)

# Jika ingin menyimpan ke file
with open('unique_categories.txt', 'w') as f:
    for category in unique_categories:
        f.write(f"{category}\n")
```
